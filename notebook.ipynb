### CRIANDO UMA NOVA COLUNA MESES COM OS MESES DO OBITO

```python
# Criando uma nova coluna chamada mes com os valores da data_obito sendo forçados a terem 8 digitos, inserindo 0 no começo dos que tem apenas 7
novo_df['mes'] = novo_df['data_obito'].astype(str).str.zfill(8)
novo_df.head()
```

```python
# Pegando o mês das datas usando o .str[2:4] para pegar o terceiro e o quarto digito da data (mês)
novo_df['mes'] = novo_df['mes'].str[2:4]
novo_df.head()
```

### Renomeando os meses

```python
meses = {
    '01': 'Janeiro',
    '02': 'Fevereiro',
    '03': 'Março',
    '04': 'Abril',
    '05': 'Maio',
    '06': 'Junho',
    '07': 'Julho',
    '08': 'Agosto',
    '09': 'Setembro',
    '10': 'Outubro',
    '11': 'Novembro',
    '12': 'Dezembro'
}
novo_df['mes'] = novo_df['mes'].replace(meses)
novo_df['mes'].value_counts()
```

### Renomeando as raças

```python
novo_df['raca'].value_counts()
```

```python
racas = {
    1: 'Branca',
    2: 'Preta',
    3: 'Amarela',
    4: 'Parda',
    5: 'Indigena'
}
novo_df['raca'] = novo_df['raca'].replace(racas)
novo_df['raca'] = novo_df['raca'].astype(str)
novo_df['raca'].value_counts()
```

### Formatando os valores de idade para uma nova coluna chamada 'anos'

```python
# Contando os valores a partir de 1 ano de idade
qtd = (novo_df['idade'] > 401).sum()
print(qtd)
```

```python
qtd = (novo_df['idade'] == 400).sum()
print(qtd)
# Provavelmente por erro, apenas um valor tem "0 anos", então não irei considerar esse 400 como 1 ano
```

```python
# Todas as idades abaixo de 400 serão considerada como 0 anos (não chegou a fazer 1 ano)
novo_df['anos'] = novo_df['idade']
mask = novo_df['anos'].between(0,400)
novo_df.loc[mask, 'anos'] = 0
```

```python
# Transforma as idades entre 401 (1 ano) e 599 (199 anos) em valores normais subtraindo 400 delas

mask = novo_df['anos'].between(401,599)
novo_df.loc[mask, 'anos'] = novo_df.loc[mask, 'idade'] - 400

novo_df.head()
```

# Criação de visualizações

```python
import plotly.express as px
```

### Proporção de óbitos registrados por raça

```python
contagem = df_final['raca'].value_counts().reset_index()
contagem = contagem[contagem['raca'] != 'Ignorado']
contagem.columns = ['raca', 'quantidade']

fig = px.pie(contagem, 
             names = 'raca',
             values='quantidade',
             title='Proporção de óbitos por Raça em 2023',
             hole=0.5)

fig.show()
```

```python
df_final['tipo_obito'].unique()
```

### Tipos de óbitos mais comuns em 2023

```python
contagem = (df_final['tipo_obito'].value_counts().reset_index())
contagem = contagem[contagem['tipo_obito'] != 'Ignorado']
contagem.columns = ['tipo_obito', 'quantidade']

fig = px.bar (
    contagem,
    x='quantidade',
    y='tipo_obito',
    orientation='h',
    title='Tipos de óbitos mais comuns em 2023',
    labels={
        'quantidade': 'Quantidade',
        'tipo_obito': 'Tipo de óbito'
    }
)

fig.update_layout(
    yaxis=dict(autorange='reversed')
)

fig.show()
```

### Meses com mais óbitos em 2023

```python
contagem = (df_final['mes'].value_counts().reset_index())
contagem.columns = ['mes', 'quantidade']

fig = px.bar (
    contagem,
    x='quantidade',
    y='mes',
    orientation='h',
    title='Meses com a maior quantidade de óbitos em 2023',
    labels={
        'quantidade': 'Quantidade',
        'mes': 'Mês'
    }
)

fig.update_layout(
    yaxis=dict(autorange='reversed')
)

fig.show()
```

### Periodo do dia com mais óbitos registrados em 2023


```python
contagem = (df_final['periodo'].value_counts().reset_index())
contagem.columns = ['periodo', 'quantidade']

fig = px.bar (
    contagem,
    x='quantidade',
    y='periodo',
    orientation='h',
    title='Periodo do dia onde mais óbitos foram registrados em 2023',
    labels={
        'quantidade': 'Quantidade',
        'periodo': 'Periodo'
    }
)

fig.update_layout(
    yaxis=dict(autorange='reversed')
)

fig.show()
```

```python
contagem = (df_final['escolaridade'].value_counts().reset_index())
contagem = contagem[contagem['escolaridade'] != 'Ignorado']
contagem.columns = ['escolaridade', 'quantidade']

fig = px.bar (
    contagem,
    x='quantidade',
    y='escolaridade',
    orientation='h',
    title='Óbitos registrados por escolaridade',
    labels={
        'quantidade': 'Quantidade',
        'tipo_obito': 'Escolaridade'
    }
)

fig.update_layout(
    yaxis=dict(autorange='reversed')
)

fig.show()
```

```python
df_filtrado = df_final[
    (df_final['tipo_obito'] == 'Suicidio') &
    (df_final['raca'] != 'Ignorado')
]


suicidios_por_raca = (
    df_filtrado[df_filtrado['tipo_obito'] == 'Suicidio']
    .groupby('raca')
    .size()
    .reset_index(name='quantidade')
)

suicidios_por_raca = suicidios_por_raca.sort_values(
    'quantidade', ascending=False
)

fig = px.bar(
    suicidios_por_raca,
    x='quantidade',
    y='raca',
    orientation='h',
    title='Quantidade de suicídios registrados por raça em 2023',
    labels={
        'quantidade': 'Quantidade de suicídios',
        'raca': 'Raça'
    }
)

fig.update_layout(
    yaxis=dict(autorange='reversed'),
    template='plotly_white'
)

fig.show()

```
